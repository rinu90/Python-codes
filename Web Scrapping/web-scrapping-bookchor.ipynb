```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
url="https://www.bookchor.com"
r= requests.get(url)
print(r)
#product-item

url="https://www.bookchor.com/category/15/99-store"
r= requests.get(url)
print(r)

soup=BeautifulSoup(r.text, "lxml")
```

```python
Price=[]
price=soup.find_all("div", class_="pi-price")
for i in price:
    price=i.text
    Price.append(price)
#print(Price)    
```

```python
Price1=[]
for j in Price:
    k=j.split(".")
    #print(k[1])
    Price1.append(k[1])
print(Price1)
print(len(Price1))
```

```python
import re
Author=[]
author=soup.find_all("p",string=re.compile("By"))
for i in author:
    author=i.text
    Author.append(author)   
```

```python
Author1=[]
Author2=[]
for j in Author:
    h=j.split(":")
    #print(h[1])
    Author1.append(h[1])
#print(Author1)    
for l in Author1:
    A=l.replace("...", "")
    #print(A)
    Author2.append(A)
print(Author2)
print(len(Author2))
```

```python
Name=[]
Name1=[]
for i in soup.find_all("h3"):
    n=i.text
    Name.append(n)
#print(Name)    

for l in Name:
    A=l.replace("...", "")
    #print(A)
    Name1.append(A)
    if 'More Books' in Name1:
        Name1.remove('More Books')
    
print(Name1)
print(len(Name1))
```

```python
import pandas as pd
data={
    'Name':Name1,
    'Price':Price1,
    'Author':Author2
}
```

```python
df=pd.DataFrame(data)
```
